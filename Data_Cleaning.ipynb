```python
# Now lets convert the month 
def Month_Converter(x):
    if pd.notnull(x):
        num1 = int(x.split(' ')[0])
        num2 = int(x.split(' ')[3])
      
        return num1*12+num2
    else:
        return x
```

```python
def get_Diff_Values_Colum(credit_column, diff_value = []):   
    column = credit_column.dropna()
    for i in column:
        if ',' not in i and i not in diff_value:
            diff_value.append(i)
        else:
            for data in map(lambda x:x.strip(), i.replace('and','').split(',')):
                if not data in diff_value:
                    diff_value.append(data)
    return dict(enumerate(sorted(diff_value)))
```

```python
def get_lower_upper(credit_column): 
    
    Q1 = credit_column.quantile(0.25)
    Q3 = credit_column.quantile(0.75)
    iqr = Q3-Q1
    
    lower_lim = Q1 - (1.5*iqr)
    upper_lim = Q3 + (1.5*iqr)
    
    return lower_lim, upper_lim
```

```python
# Reassign Wrong Values and Show Function
def Reassign_Group_Numeric_Wrong_Values(credit, groupby, column, inplace=True):    

    # Identify Wrong values Range
    def get_group_min_max(credit, groupby, column):            
        cur = credit[credit[column].notna()].groupby(groupby)[column].apply(list)
        x, y = cur.apply(lambda x: stats.mode(x)).apply([min, max])
        return x[0][0], y[0][0]
    
    # Assigning Wrong values
    def make_group_NaN_and_fill_mode(credit, groupby, column, inplace=True):
        credit_dropped = credit[credit[column].notna()].groupby(groupby)[column].apply(list)
        x, y = credit_dropped.apply(lambda x: stats.mode(x)).apply([min, max])
        mini, maxi = x[0][0], y[0][0]

 # assign Wrong Values to NaN
        col = credit[column].apply(lambda x: np.NaN if ((x<mini)|(x>maxi)) else x)

        # fill with local mode
        mode_by_group = credit.groupby(groupby)[column].transform(lambda x: x.mode()[0] if not x.mode().empty else np.NaN)
        result = col.fillna(mode_by_group)

        # inplace
        if inplace:
            credit[column]=result
        else:
            return result
 # Run      
    if inplace:   
        print("\nExisting Min, Max Values:", credit[column].apply([min, max]), sep='\n', end='\n')       
        mini, maxi = get_group_min_max(credit, groupby, column)        
        print(f"\nby {groupby}'s Actual min, max Values:", f'min:\t{mini},\nmax:\t{ maxi}', sep='\n', end='\n')
        
        # Before Assigning Wrong values      
        x = credit[column].value_counts(dropna=False).head(1)
        print(f'\nBefore Assigning Min, Max {column}:', [x if x.index[0] is np.NaN or x.index[0] is None else 'No NaN Value'] , end='\n')
        
        a = credit.groupby(groupby)[column].apply(list) 
        print(f'\nBefore Assigning Example {column}:\n', *a.head().values, sep='\n', end='\n')

 # Assigning
        make_group_NaN_and_fill_mode(credit, groupby, column, inplace)
        
        # After Assigning Wrong values
        y = credit[column].value_counts(dropna=False).head(1)
        print(f'\nAfter Assigning  Min, Max {column}:', [y if y.index[0] is np.NaN or x.index[0] is None else 'No NaN Value'] , end='\n')
        
        b = credit.groupby(groupby)[column].apply(list)
        print(f'\nAfter Assigning Example {column}:\n', *b.head().values, sep='\n', end='\n')
    else:   
        # Show
        return make_group_NaN_and_fill_mode(credit, groupby, column, inplace)
```
